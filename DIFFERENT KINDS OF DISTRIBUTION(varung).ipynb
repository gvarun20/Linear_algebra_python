# Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Set seed for reproducibility
np.random.seed(42)

# Generate uniform distribution
uniform_distribution = np.random.uniform(low=data.min(), high=data.max(), size=data.shape)

# Plot the original dataset and the uniform distribution
plt.figure(figsize=(12, 6))

# Plot Original Data
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=iris.target, cmap='viridis', edgecolors='k', s=50)
plt.title('Original Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Plot Uniform Distribution
plt.subplot(1, 2, 2)
plt.scatter(uniform_distribution[:, 0], uniform_distribution[:, 1], c='red', edgecolors='k', s=50)
plt.title('Uniform Distribution for Iris Dataset')
plt.xlabel('Random Feature 1')
plt.ylabel('Random Feature 2')

plt.tight_layout()
plt.show()
```

# normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Set seed for reproducibility
np.random.seed(42)

# Generate normal distribution
mean = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
normal_distribution = np.random.normal(loc=mean, scale=std_dev, size=data.shape)

# Plot the original dataset and the normal distribution
plt.figure(figsize=(12, 6))

# Plot Original Data
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=iris.target, cmap='viridis', edgecolors='k', s=50)
plt.title('Original Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Plot Normal Distribution
plt.subplot(1, 2, 2)
plt.scatter(normal_distribution[:, 0], normal_distribution[:, 1], c='blue', edgecolors='k', s=50)
plt.title('Normal Distribution for Iris Dataset')
plt.xlabel('Random Feature 1')
plt.ylabel('Random Feature 2')

plt.tight_layout()
plt.show()
```

# binomial distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Number of trials (n) and probability of success (p)
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate binomial distribution
binomial_distribution = np.random.binomial(n, p, size=1000)

# Plot the histogram of the binomial distribution
plt.hist(binomial_distribution, bins=np.arange(0, n + 2) - 0.5, density=True, edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(np.arange(0, n + 1))
plt.show()
```

# Poisson distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Average rate of events per interval (lambda)
lambda_param = 3.0

# Generate Poisson distribution
poisson_distribution = np.random.poisson(lambda_param, size=1000)

# Plot the histogram of the Poisson distribution
plt.hist(poisson_distribution, bins=np.arange(0, max(poisson_distribution) + 2) - 0.5, density=True, edgecolor='black', alpha=0.7)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.xticks(np.arange(0, max(poisson_distribution) + 1))
plt.show()
```

# Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Probability of success (p)
p = 0.3

# Number of samples
num_samples = 1000

# Generate Bernoulli distribution
bernoulli_distribution = np.random.choice([0, 1], size=num_samples, p=[1 - p, p])

# Plot the histogram of the Bernoulli distribution
plt.hist(bernoulli_distribution, bins=[-0.5, 0.5, 1.5], density=True, edgecolor='black', alpha=0.7)
plt.title('Bernoulli Distribution')
plt.xlabel('Outcome (0: Failure, 1: Success)')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.show()
```

# log-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters for the log-normal distribution
mean_log = 0.0  # Mean of the natural logarithm
std_dev_log = 0.2  # Standard deviation of the natural logarithm

# Number of samples
num_samples = 1000

# Generate log-normal distribution
log_normal_distribution = np.random.lognormal(mean_log, std_dev_log, size=num_samples)

# Plot the histogram of the log-normal distribution
plt.hist(log_normal_distribution, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```
